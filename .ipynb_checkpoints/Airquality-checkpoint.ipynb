<p style='text-align: justify;'> 
The COVID-19 pandemic and subsequent lockdowns had a significant impact on air quality in major cities worldwide. With reduced industrial activities and limited mobility, emissions from vehicles and factories decreased, leading to improved air quality. This resulted in lower levels of pollutants such as nitrogen dioxide and particulate matter, leading to health benefits and better quality of life for residents. Its data is vital for understanding air quality, supporting policy-making, and conducting climate change and environmental research..</p>

## 3. Visualisation of NO2 during March 2019 & post covid
 [Go back to the "Table of contents"](#Table-of-contents)
 

Now, we are going to create maps of the NO2 column over study area:
* The first map will depict the averaged NO2 column over the study area for the entire period before the COVID-19     lockdown in 20**19** 
* It will be compared with the average NO2 column during and after the lockdown period in the area of interest

This process will be implemented using the atmospheric toolbox's 'merge' operation. However, before carrying out the operation, we need to remove unwanted variables to reduce the processing time during the merge operation.

The 'keep' operation can be used to select the variables of interest from the entire dataset. This step is more time-efficient and reduces the size of variables and data that need to be processe

Final 'bin' operation can be used to find the temporal average of all of the Level3 (L3) products d.

```python
#where the S5P files are stored
input_path= './Processing/batch_download/'
#where the pre-processed files will be saved
export_path='./Processing/pre_processed_files/'
for region in city:
    #roi=extent[region]
    for date in year_list:
        # the function `list` recovers the name of all files in the input_path folder
        files_list =glob.glob(join(input_path,str(region),str(date),'*.nc'))
        # Make sure that list_files contains only S5P files (filter the list files)
        files_list=list(filter(lambda x:os.path.basename(x).startswith("S5P_OFFL_"), files_list))
        #sort files acoording to time 
        files_list=sorted(files_list, key=lambda x:os.path.basename(x).split('____')[-1][:8])
        
        #Starting time of the preprocessing
        t0 = time.time()
        
        #Pre-processing loop on each file in input_path to get rid of unnecessary variables (keep()) and crop to our ROI
        for i,f in enumerate(files_list):
            file_name=os.path.basename(f)
            #Pre-processing
            try:
       
                Converted_NO2=harp.import_product(f, \
                                                  
                operations="keep(latitude,latitude_bounds,longitude,longitude_bounds, \
                tropospheric_NO2_column_number_density,tropospheric_NO2_column_number_density_validity)")
              

                #Export of the preprocessed file to export_path
                harp.export_product(Converted_NO2, join(export_path,region,str(date), file_name),file_format="netcdf") 
            
            except Exception as err:
                #raise (err)
                print (err)

            else:
            
                print("product", file_name,"pre-processed")
        
        #End time of the pre-precessing
        t1 = time.time()
        
        pre_processing_time = t1-t0
        
        print("Pre-processing time: {} seconds".format(pre_processing_time))

       
```

```python
extent
```

Let's now merge the pre-processed products together to create a 2 week averaged NO2 product.

```python
# Specify input and output of harp.import_product
#All pe-processed S5P L2 NO2 product for 2019
lat_edge_step=0.01
lon_edge_step=0.01
for region in city:
  #Start time of the processing
    t0 = time.time()
    for year in year_list:
        try:
            input_files=f'./Processing/pre_processed_files/{region}/{year}/S5P_OFFL_L2__NO2_*.nc' 
            #output file
            export_file=f'./Processing/merged_files/{region}/converted_NO2_{year}_final.nc' 
            #Use harp.import_product python function to generate the merged product
          
            #Merging of the pre-processed files
            temp=extent[region]
            lat_edge_length, lat_edge_offset=int((temp[3]-temp[1])//lat_edge_step),temp[1]
            lon_edge_length, lon_edge_offset=int((temp[2]-temp[0])//lat_edge_step),temp[0]
            if lat_edge_length<=1 :
                lat_edge_length=int(lat_edge_length+1)
            if lon_edge_length<=1 :
                lon_edge_length=int(lon_edge_length+1)
           
            pre_op= "tropospheric_NO2_column_number_density_validity>75;" \
            + f"bin_spatial({lat_edge_length}, {lat_edge_offset}, {lat_edge_step}, {lon_edge_length}, {lon_edge_offset}, {lon_edge_step});" \
            + "derive(latitude {latitude}); derive(longitude {longitude});\
            keep(latitude,longitude,tropospheric_NO2_column_number_density,weight)"
    
        
            post_op="bin(); squash(time, (latitude,longitude,tropospheric_NO2_column_number_density))"
              
            Converted_NO2 = harp.import_product(input_files,operations=pre_op,post_operations= post_op)
         
    
            #Export of the merged file        
            harp.export_product(Converted_NO2, export_file,file_format="netcdf")
        except Exception as err:
            print(err)
        

    #End time of the processing
    t1 = time.time()

    processing_time = t1-t0

    print("Processing time for {}: {} seconds".format(region,processing_time))
   
```

The attributes of the averaged NO2 products are as follows:
* final grid:
   The retrieved product is transformed from the time-binned domain into a two-dimensional space domain. The gridded product is specified with a cell resolution of 0.01° by 0.01°.
* We have applied a filter to exclude observations with a quality flag below 75. This threshold value aligns with the recommendation stated in the <a href="https://sentinel.esa.int/documents/247904/2474726/Sentinel-5P-Level-2-Product-User-Manual-Nitrogen-Dioxide"> Sentinel-5P NO2 Product User Manual </a>. The quality flag is determined based on various factors, including cloud cover, surface albedo, presence of snow-ice, saturation, and geometry. These factors are taken into consideration when defining the "quality assurance value" (qa_value) for each individual observation, which allows users to easily filter out less accurate data.

The qa_value is a continuous variable ranging from 0 (indicating an error) to 1 (representing an optimal retrieval)
s.

## 3.2. Visual Annual comparison of NO2 

Now that we have created the merged products comprising the pre, during, and post-lockdown years, we can proceed to map the averaged NO2 column from each respective year.

Before creating the maps, we will define a function named `figures()`.

This function  takes as input the following arguments:
* data: the data we want to visualize (must be an `xr.DataArray`)
* ax: the axe used for the figure (a `plt.subplot` with a defined projection)
* title: title of the figure (`string`)
* min_value: data min value for the colorbar (`float`)
* max_value: data max value for the colorbar (`float`)
* hv: orientation of the colorbar (horizontal or vertical) (`string`)
We will use the `pcolormesh()` function for the colour mapping.


```python
def figures(data:xr.DataArray,ax:plt.subplot,title:str,min_value:float, max_value:float,hv:str) -> plt.subplot:
    """
    Creates a coloured map of NO2 tropospheric column. 
    Args:
        (xarray.dataArray) data, NO2 tropospheric column
        (subplot) ax, supblot with a defined projection
        (str) title, title of your figure
        (float) min_value, data min value for the colorbar
        (float) max_value, the max value of the data for the colorbar 
        (str) hv, the orientation of the colorbar : horizontal or vertical
    Returns:
        (subplot) coloured map of the NO2 tropospheric column
    """
        
    
    # Coloured Map of O3 total column levels
    im1=data.plot.pcolormesh(ax=ax, x='longitude', y='latitude',add_colorbar=False, shading='flat',
                            cmap='jet', transform=ccrs.PlateCarree(), vmin=min_value, vmax=max_value)
    
    # Figure settings
    ax.set_title(title ,fontsize=25)
    ax.add_feature(cartopy.feature.LAND, edgecolor='black')
    ax.add_feature(cartopy.feature.OCEAN)
    ax.add_feature(cartopy.feature.RIVERS)
    ax.coastlines('10m',linewidth=2)
    ax.gridlines()
    cbar = plt.colorbar(im1,ax=ax, orientation=hv,pad=0,shrink=0.8)#pad adjust the space between plots and colorbar
    cbar.ax.tick_params(labelsize=20)
    
    return (ax)
```

```python
# Directory where the data is stored
merged_dir='./Processing/merged_files/'
```

```python
#comparison for Paris

#first for 2019 avearged No2
Fname2019=glob.glob(join(merged_dir,city[0],"*2019*.nc"))
#open the netcdf dataset with x-array
FIn2019 = xr.open_dataset(Fname2019[0])
#susbset data to 2019 and rescale with *10^3
VarNO22019=FIn2019['tropospheric_NO2_column_number_density']*(10**3)


#second , 2020 avearge No2
Fname2020=glob.glob(join(merged_dir,city[0],"*2020*.nc"))
#open the netcdf dataset with x-array
FIn2020 = xr.open_dataset(Fname2020[0])
#susbset data to 2019 and rescale with *10^3
VarNO22020=FIn2020['tropospheric_NO2_column_number_density']*(10**3)
```

The following cell creates one figure with two subplots on the same row (one subplot per year)

```python
fig,(axe1,axe2)=plt.subplots(1,2,figsize=(30,10)) #1 row, 2 columns
vmin=0
axe1 = plt.subplot(121, projection=ccrs.Orthographic(11,42))
title = 'Pre lockdown in Paris (2019)'
figures(VarNO22019,axe1,title,vmin,VarNO22019.max(),"horizontal")

axe2 = plt.subplot(122, projection=ccrs.Orthographic(11,42))
title = 'During lockdown in Paris (2020)'
figures(VarNO22020,axe2,title,vmin,VarNO22020.max(),"horizontal")

fig.suptitle('S-5p averaged NO$_2$ tropospheric column (mmol.m-2)',fontsize=35,va='bottom')

# Save figure to file
plt.savefig('./Processing/figures/NO2_2019&2020', facecolor='white',\
            bbox_inches='tight', dpi=300)
```

## 3.3. Show number of averaged timestep

In this section, we will analyze the number of measurements used to calculate the averaged NO2 column for each pixel. We will do this after filtering out the pixels with a quality index lower than 0.75, which can be attributed to factors like the presence of cloud contamination.

The variable called 'weight,' computed either by 'harpmerge' or 'harp.import_product,' tracks the number of L2 pixels utilized in calculating the average within each grid bin. A higher 'weight' value for a bin indicates a more reliable averaged value for that particular bin. This variable serves as a valuable indicator of data confidence.

```python
# Assign  name to the respective annual average weight
Weight2019=FIn2019['weight']

Weight2020=FIn2020['weight']
```

```python
# subplot with two horizontal plot
fig,(axe1,axe2)=plt.subplots(1,2,figsize=(30,10)) #1 row, 2 columns
vmin=0
axe1 = plt.subplot(121, projection=ccrs.Orthographic(11,42))
title = 'QA_value (weight) for 2019'
figures(Weight2019[0],axe1,title,vmin,Weight2019[0].max(),"horizontal")

axe2 = plt.subplot(122, projection=ccrs.Orthographic(11,42))
title = 'QA_value (weight) for 2020'
figures(Weight2020[0],axe2,title,vmin,Weight2020[0].max(),"horizontal")

fig.suptitle('S-5p number of measures used to compute each averaged NO$_2$ column bin',fontsize=35,va='bottom')

# Save figure to file
plt.savefig('./Processing/figures/NO2_2019&2020_QA', bbox_inches='tight', dpi=300)
```

## 4.2. Time series in Paris

This exercise was inspired by <a href="https://www.esa.int/Applications/Observing_the_Earth/Copernicus/Sentinel-5P/Air_pollution_in_a_post-COVID-19_world"> an ESA article </a>. We are going to reproduce a time series of NO2 concentration over Paris. The studied area is highlighted in orange in the below figure and is delimited by the following geographical coordinates: 
* Lat: 48.59/49.11° 
* Lon: 1.88/2.74° 

The time period lies between February and June 2020. We are also going to apply the same processing to year 2019 to be able to assess the effects of the lockdown.

<img src=./graphics/NO2_timeseries_Paris.png width='810'/>

To compute the averaged NO2 tropos. column on one day, we average all S5P pixels falling in the delimited square (around the city of Paris) and over 14 consecutive days. It means that, for example, the value that we will compute on March 5th will be the averaged NO2 between March 5th to 18th, the value on March 6th will be the average from March 6th to March 19th, and so on. We are computing a sliding average over our time period.


### 4.2.1. File concatenation

In this part, we compute the average over the defined area. This area is considered as a single bin. We will only work on 2020 (the same operations on 2019 have already been carried out).


```python
# Specify input and output file
input_path='/shared/S5P_NO2_092021/training_NO2_23092021/Processing/pre_processed_files/Paris_2020/' #All S5P L2 NO2 product in 2020
export_path='/shared/S5P_NO2_092021/training_NO2_23092021/Processing/merged_files/Time_series_Paris_2020/' #NetCDF file where merged product will be stored
```

We create a list of all input paths:

```python
# Check the name of all files in the folder and sort the file in alphabetic name
files_input= sorted(glob.glob(join(input_path, 'S5P_OFFL*.nc')))
```

The following cell :
 - takes as input the variable created above `files_input` and operates a sliding average over 14 consecutive days `files_input[i:i+14]`
 - Determines the box where the average is computed, here (lat: 48.59/49.11° ; lon: 1.88/2.74°)
 - does not perform a bin_spatial operation because we consider the area as a single bin
 - We perform a `derive(tropospheric_NO2_column_number_density [Pmolec/cm2])` operation to convert NO2 values from mol/cm2 to Pmol/m2

```python
#Use harp.import_product python function to generate the averages

#Define the coordinates of our ROI
lon_min=1.88
lon_max=2.74
lat_min=48.59
lat_max=49.11

#Define the operation list to be applied to each product before being merged
op='latitude >='+str(lat_min)+'[degree_north] ; latitude <='+str(lat_max)+'[degree_north] ; \
    longitude<='+str(lon_max)+'[degree_east];longitude>='+str(lon_min)+'[degree_east];\
    tropospheric_NO2_column_number_density_validity>75; \
    keep(latitude,longitude,tropospheric_NO2_column_number_density)'

for i in range(len(files_input)-14):
    Converted = harp.import_product(files_input[i:i+14], \
                      operations= op, \
                      post_operations="bin();derive(tropospheric_NO2_column_number_density [Pmolec/cm2]); squash(time, (latitude,longitude))")
           
    harp.export_product(Converted, join(export_path, 'converted_{:0>3d}.nc'.format(i)) ,file_format="netcdf")
```

Both cells below, retrieve the NO2 value for each 14 day average and integrate it into a variable. We end up with an array of the values of the sliding NO2 average for the consecutive dates. Moreover, we define the same operation for the time variable.

For 2019, the analysis has already been carried out the same way as for 2020. The cell above has also been run for 2019, and all output files are in the `/shared/S5P_NO2_092021/training_NO2_23092021/Processing/merged_files/Times_series_Paris_2019/`. The date values for year 2019 have been saved to an other netcdf file `time_Paris_2019.nc`

```python
# Gather the averaged values for each 2019 day in a variable
average_2019=[] #create an empty array for 2019 NO2 average

files_output_2019= sorted(glob.glob('/shared/S5P_NO2_092021/training_NO2_23092021/Processing/merged_files/Times_series_Paris_2019/conver*.nc'))


for i in range(len(files_output_2019)):
    FIn=xr.open_dataset(files_output_2019[i])
    average_2019.append(FIn.tropospheric_NO2_column_number_density.data)
    FIn.close()

# Open the netcdf file of time 2019  
FIn=xr.open_dataset('/shared/S5P_NO2_092021/training_NO2_23092021/Processing/merged_files/Times_series_Paris_2019/time_Paris_2019.nc')
time_2019=FIn.time.data 
```

```python
time_2019
```

For 2020, the analysis has been carried out in this notebook, and the cell below recovers the outputs.

```python
 # Gather the average values for each 2020 day in a variable
average_2020=[]
time_2020=[]

files_output_2020= sorted(glob.glob('/shared/S5P_NO2_092021/training_NO2_23092021/Processing/merged_files/Time_series_Paris_2020/conver*.nc'))
files_input_2020=sorted(glob.glob('/shared/S5P_NO2_092021/training_NO2_23092021/Processing/pre_processed_files/Paris_2020/S5P*.nc'))

for i in range(len(files_output_2020)):
    FIn=xr.open_dataset(files_output_2020[i])
    average_2020.append(FIn.tropospheric_NO2_column_number_density.data)
    FIn.close()
    
# For the variable `datetime_start` we only keep the first value because it 
# is a time vector of each pixel of the image. We only want the sensing start time of the image
for i in range(len(files_input_2020)-14):
    FIn=xr.open_dataset(files_input_2020[i])
    time_2020.append(FIn.datetime_start.data[0]) 
    FIn.close()
```

### 4.2.2. Time series plot

In the cell below, we plot both time series: 2019 and 2020

```python
fig,(axe1,axe2)=plt.subplots(1,2, figsize=(16,8))
color0='r'
color1='b'

#---------plot time evolution of NO2 2019-----------#
axe1.plot(time_2019,average_2019,color0) 
axe1.set_ylabel('NO2 2019 [Pmolec/cm²]', fontsize=14)
axe1.set_ylim(0, 12)
axe1.set_title('S-5p L2 NO$_2$ - Paris - February-June 2019',fontsize=17) # add title
# put the date in the correct format on x
date_form = DateFormatter("%d-%m")
axe1.xaxis.set_major_formatter(date_form)

#---------plot time evolution of NO2 2020-----------#
axe2.plot(time_2020,average_2020,color1) 
axe2.set_ylabel('NO2 [Pmolec/cm²]', fontsize=14)
axe2.set_ylim(0, 12)
axe2.set_title('S-5p L2 NO$_2$ - Paris - February-June 2020 ',fontsize=17) # add title
# put the date in the correct format on x
date_form = DateFormatter("%d-%m")
axe2.xaxis.set_major_formatter(date_form)

# Save the plot
plt.savefig('/shared/S5P_NO2_092021/training_NO2_23092021/Processing/figures/NO2_time _series_Paris.png', facecolor='white',\
            bbox_inches='tight', dpi=300)
```

We note that in 2020 the NO2 concentrations over Paris have decreased compared to the same period in 2019. In 2019, there is a significant but brief decrease at the end of February. Then in March 2020, the NO2 levels are significantly and durably reduced due to the lockdown. 

This figure highlights a drastic decrease in NO2 emission over Paris in Spring 2020 compared to the same period in 2019. The following cells make it possible to quantify this reduction:


```python
# Compute the average of NO2 during all period (5 months)
global_average_2019=np.mean(average_2019)
global_average_2020=np.mean(average_2020)

# print values
print( 'Mean NO2 tropospheric column over Paris in 2019 is: {} Pmolec/cm²'.format(global_average_2019))
print( 'Mean NO2 tropospheric column over Paris in 2020 is: {} Pmolec/cm²'.format(global_average_2020))
```

Now we quantify the percentage of reduction:

```python
diff_NO2=(global_average_2020-global_average_2019)/global_average_2019*100

print('The NO2 level decrease in 2020 compared to 2019 over Paris is: {:.3}%'.format(diff_NO2))
```

The negative value shows a strong decrease diminution of the emission of NO2.
