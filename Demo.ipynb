# 1. Introduction
 [Go back to the "Table of contents"](#Table-of-contents)

The fusion of Optical and Synthetic Aperture Radar (SAR) data has emerged as a powerful and innovative technique in the field of remote sensing and Earth observation. Optical sensors provide high-resolution and detailed imagery, capturing information about visible and near-infrared wavelengths. On the other hand, SAR sensors offer all-weather and day-night imaging capabilities, measuring the backscatter of microwave signals. By combining these complementary datasets, the fusion process enhances the overall information content, enabling researchers and analysts to unlock a wealth of valuable insights and applications in areas such as land cover classification, change detection, urban monitoring, disaster assessment, and environmental studies. This integration of Optical and SAR data holds great promise for advancing our understanding of the Earth's surface and addressing various scientific and societal challenges.


<strong>Sentinel 1: </strong>

Sentinel-1, launched in April 2014, is part of the European Space Agency's Copernicus program. It is a versatile radar imaging satellite that operates day and night and in all weather conditions, making it ideal for continuous Earth observation. Sentinel-1 carries a synthetic aperture radar (SAR) instrument, which emits microwave signals and measures the backscatter to capture high-resolution images of the Earth's surface. With its ability to penetrate clouds and capture data regardless of weather conditions, Sentinel-1 is particularly valuable for monitoring changes in ice, land, and ocean surfaces, as well as for disaster management and maritime surveillance. It plays a crucial role in various applications, including detecting oil spills, monitoring floods, tracking land subsidence, and supporting agricultural activities, contributing significantly to global monitoring efforts and informed decision-making.</b> More information on the characteristics of Sentinel-1 can be found <a href="https://sentinel.esa.int/web/sentinel/user-guides/sentinel-1-sar">>>> Here! </a>

<strong>Sentinel 2: </strong>

Sentinel-2, launched in June 2015, is part of the European Space Agency's Copernicus program. It consists of a constellation of twin satellites, Sentinel-2A and Sentinel-2B, operating in tandem to provide systematic and high-resolution multispectral imagery of the Earth's surface. Equipped with a multi-spectral imaging instrument, Sentinel-2 captures data across 13 spectral bands, ranging from visible to infrared wavelengths, enabling detailed monitoring of land cover, vegetation health, and changes in the environment. With its frequent revisits and global coverage, Sentinel-2 plays a crucial role in various applications, including agriculture, forestry, urban planning, disaster monitoring, and environmental management, contributing significantly to the advancement of Earth observation and sustainable development.</b> More information on the characteristics of Sentinel-2 can be found <a href="https://sentinel.esa.int/web/sentinel/user-guides/sentinel-2-msi">>>> Here! </a>

## 3.2 Preprocessing 
 [Go back to the "Table of contents"](#Table-of-contents)
 
 
The function conducts preprocessing operations on the stacked time series image, including NA pixel masking based on a threshold value.
The pixels with values above the specified threshold are considered to have sufficient information and reliability for the time series analysis and are retained, 
while pixels with values below the threshold are masked out or excluded from further processing.

This threshold can be dynamically estimated using the median value of non-NA pixel counts or specified as a static threshold in percentage units (e.g., 85%).
When the dynamic method is selected, the threshold should be set to None, and vice versa for the static method.Finally, if the interpolation parameter is set to True, 
the function performs interpolation for the retained pixels in the time series data.

```python

print("----------------------------------------------------------------\n \
                 preprocessing s2_images\n \
---------------------------------------------------------------------\n")
#defualt  preprocessing parameters 
processes={"mask":True, "method":"median", "tolerance":0,\
           
            "threshold":None, "interpolate":True}


with ProgressBar():          
    svi_dates,svi_prep=preprocess_s2(image_stack=image_stack,param=processes)
```

## 4.2 Preprocessing
 [Go back to the "Table of contents"](#Table-of-contents)

It is highly recommended to convert backscattered data from decibels (dB) to power intensity before conducting any numerical analysis. This transformation ensures that the data is in a linear scale, allowing for more accurate and meaningful interpretations and computations.

The clip_percentile function plays a crucial role in removing outliers from the data by setting upper and lower bounds based on the histogram distribution. This function identifies extreme values in the data that lie beyond a certain percentile range and clips (trims) them to the specified bounds. By setting these bounds, extreme values that might be erroneous or influential to the analysis are restricted, ensuring a more reliable and consistent dataset.However, the clip_percentile step is optional, and it can be skipped to reduce the processing time.
 

```python
print("----------------------------------------------------------------\n\
                convert db to power and removing outliers\n\
---------------------------------------------------------------------\n")


s1_polz_pwr=[]
for i in tqdm(range(len(s1_polz)),desc='db-pwr-rm_outlier',colour='blue'):
    with ProgressBar():
        s1_pwr=db_to_pwr(s1_polz[i])
        ## remove extreme value
        s1_polz_pwr.append(clip_percentile(s1_pwr,lb=2,ub=98, axis=0))
        #s1_polz_pwr.append(s1_polz[i])
        #closing the data after appending to free the memory
       
        s1_polz[i]=None
#----------------------------------------------------------------
                 
                      #visualization 
#---------------------------------------------------------------- 
    

viz=True 
if viz:
    # for n,>1,>1 time series
    if s1_polz_pwr[0].shape[1]>1 or s1_polz_pwr[0].shape[2]>1 :
        kk=pd.Series(np.nanmean(s1_polz_pwr[0].compute(),axis=(1,2)))
    #for n,1,1 times series
    else :
        k_array=s1_polz_pwr[0].compute()
        kk=pd.Series(k_array.data.ravel())  

    kk.plot(figsize=(16,8))
    plt.show()
    
```

## 6. Fusion of optical and sar timeseries 
 [Go back to the "Table of contents"](#Table-of-contents)

The fusion process is carried out using the fusion_AVE module, where 'AVE' stands for average. This method employs a weighted averaging technique to fuse optical and SAR data at the pixel level. 
The fusion weight can either be pre-defined or optimized using the optuna package. The optimal weight is determined based on various available metrics in the skimage.metric package.

The function takes several parameters, which are as follows:
* input image: a tuple holding the list of images to be fused.This tuple can either be an already assigned variable containing the image objects, or it can be a path pointing to the directory where the images are stored. For large-scale fusion operations, it is recommended to use the path option to efficiently handle and process the images.
* obs_datelist: a list of observation dates in plain string(e.g '20170510'), which represents the list of observed time steps of optical data for the entire time series (observed + interpolated). These dates are resampled to the nearest 6-day resolution, and their corresponding optical images are equally weighted as the SAR image without any optimization.
* fusion_param which composed of :
    * priority_ids: A list of integers representing time steps in the time series that must be averaged with equal weights of optical and SAR data.
    * seed: The seed used for initializing the generated weight in the optuna package.
    * saveproduct: A boolean flag indicating whether the final fused products should be saved.
    * geotrans: Geotransformation to be used for writing the fused raster product in GeoTIFF format.
    * proj: The projection for writing out the fused product.
    * prefix: An optional prefix to be concatenated with the output name of the fused product.
    * nworkers: The number of cores required to perform the fusion processing in parallel.
    * optimization: If the optimize is set to True , then optimization will be performed otherwise a pre-defined weight must be provided.A variable that includes weight bounds specifying the range of weights to search from (usually between 0 and 1). It also       includes opt_metric, which specifies the quality assessment metric to be used during the fusion process. It can take values from the                   skimage.metric package, such as 'psnr', 'nmi', 'nrmse', 'mse', 'ssim', or a combination of these ('all').

<div class="alert alert-block alert-info">
<b>Tip:</b>
Preliminary tests shows that MSE (Mean square error) and PSNR (Peak Signal-to-Noise Ratio) provide more accurate optimized results.
</div>

* weight: If the weights are to be pre-defined without optimization, equal weights (0.5, 0.5) are assigned to observed time steps of optical and SAR data, while specified weights are assigned to interpolated optical and observed SAR time steps in the time series. For example, a higher weight (0.6) may be assigned to observed SAR data, while a lesser weight (0.4) to the interpolated optical SAR time steps. Note that for pre-defined weight, the sum of the weight list must be equal to 1.
